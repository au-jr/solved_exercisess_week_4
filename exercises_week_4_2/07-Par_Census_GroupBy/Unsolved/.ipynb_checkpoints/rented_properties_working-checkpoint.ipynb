```python
# Dependencies
from pathlib import Path
import pandas as pd
```

```python
# Save file path to variable
path = Path('Resources/2016_rental_by_LGA.csv')
```

```python
# Read with Pandas

df = pd.read_csv(path)
df.head()
```

```python
# Create a groupby object by state
df_state_gb = df.groupby(['State'])
df_state_gb
```

```python
# Create a dataframe of the totals for each state
state_total_df = df_state_gb.sum()
state_total_df.head()
```

```python
state_total_df.sum()
```

```python
# Calculate the total rented properties in each state by summing all of the columns together.
# Place the results into a new column.
state_total_df['Grand_Total'] = state_total_df.sum(axis=1)
state_total_df['Grand_Total']
```
