```python
# Import Dependencies
from pathlib import Path
import pandas as pd
```

```python
# Make a variable reference to the file where the CSV is located
path = Path('../Resources/dga_lpg.csv')
# Import the data into a Pandas DataFrame
df = pd.read_csv(path)
```

```python
# Reduce to columns that are in English
df.head(2)
df.columns

cols = ['Part', 'Type', 'Type.1', 'Brand', 'Model', 'Other Information','Place of Manufacture', 'Applicant', 'Telephone Number', 'Approval Expiry Date']
df_sorted = df.loc[:, cols]
```

```python
# look for missing values
df_sorted.count()
```

```python
# Drop null rows
# df_sorted = df_sorted.dropna()
df_sorted.dropna(inplace=True)
```

```python
# Verify counts
df_sorted.count()
```

```python
# List unique values of "Applicant" to locate any that may be the same
df_sorted['Applicant'].unique()
```

```python
# Combine similar applicants together
df_combined_electrolux = df_sorted.loc[df_sorted['Brand'] == 'Electrolux']
```

```python
# Check to see if you combined similar applicants correctly in "Applicant"
df_combined_electrolux['Brand'].unique()
```
